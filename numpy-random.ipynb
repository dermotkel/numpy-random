```python
import numpy as np
%matplotlib inline
```

```python
x = np.random.rand(10000)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(x)
plt.show()
```

```python
s = np.random.uniform(-1,0,1000)
```

```python
plt.hist(s)
plt.show()
```

```python
p = np.random.normal(0, 5, 10000)
```

```python
plt.hist(p)
plt.show()
```
